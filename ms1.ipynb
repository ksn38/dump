```python
import pandas as pd
import numpy as np
```

```python
SP = pd.read_csv('data/^GSPC.csv').dropna()
SP = SP[['Date', 'Close']][-5350:]
SP.columns = ['date_added', 'gspc']
SP.head()
```

```python
VIX = pd.read_csv('data/^VIX.csv').dropna()
VIX = VIX[['Date', 'Close']]
VIX.columns = ['date_added', 'vix']
VIX.head()
```

```python
TR = pd.read_csv('data/^TNX.csv').dropna()
TR = TR[['Date', 'Close']]
TR.columns = ['date_added', 'tnx']
TR.head()
```

```python
GDAXI = pd.read_csv('data/^GDAXI.csv').dropna()
GDAXI = GDAXI[['Date', 'Close']]
GDAXI.columns = ['date_added', 'gdaxi']
GDAXI.head()
```

```python
NS = pd.read_csv('data/^IXIC.csv').dropna()
NS = NS[['Date', 'Close']]
NS.columns = ['date_added', 'ixic']
NS.head()
```

```python
RU = pd.read_csv('data/^RUT.csv').dropna()
RU = RU[['Date', 'Close']]
RU.columns = ['date_added', 'rut']
RU.head()
```

```python
SS = pd.read_csv('data/000001.SS.csv').dropna()
SS = SS[['Date', 'Close']]
SS.columns = ['date_added', 'ss']
SS.head()
```

```python
BS = pd.read_csv('data/^BSESN.csv').dropna()
BS = BS[['Date', 'Close']]
BS.columns = ['date_added', 'bsesn']
BS.head()
```

```python
SZ = pd.read_csv('data/399001.SZ.csv').dropna()
SZ = SZ[['Date', 'Close']]
SZ.columns = ['date_added', 'sz']
SZ.head()
```

```python
x = pd.merge(SP, VIX, how='left', on='date_added')
x = pd.merge(x, TR, how='left', on='date_added')
x = pd.merge(x, GDAXI, how='left', on='date_added')
x = pd.merge(x, NS, how='left', on='date_added')
x = pd.merge(x, RU, how='left', on='date_added')
x = pd.merge(x, BS, how='left', on='date_added')
x = pd.merge(x, SS, how='left', on='date_added')
x = pd.merge(x, SZ, how='left', on='date_added')
```

```python
x.vix = x.vix.fillna(method='ffill')
x.vix = x.vix.fillna(method='bfill')
x.tnx = x.tnx.fillna(method='ffill')
x.tnx = x.tnx.fillna(method='bfill')
x.gdaxi = x.gdaxi.fillna(method='ffill')
x.gdaxi = x.gdaxi.fillna(method='bfill')
x.ixic = x.ixic.fillna(method='ffill')
x.ixic = x.ixic.fillna(method='bfill')
x.rut = x.rut.fillna(method='ffill')
x.rut = x.rut.fillna(method='bfill')
x.bsesn = x.bsesn.fillna(method='ffill')
x.bsesn = x.bsesn.fillna(method='bfill')
x.ss = x.ss.fillna(method='ffill')
x.ss = x.ss.fillna(method='bfill')
x.sz = x.sz.fillna(method='ffill')
x.sz = x.sz.fillna(method='bfill')
x.head()
```

```python
# x.to_csv('mybl_ticker.csv', header=False, index=False) #для косольного ввода
x.to_csv('mybl_ticker.csv', index_label='id')
```
