```python
ms1 = pd.read_csv('data/mybl_ticker.csv', index_col=0)
ms1.describe()
```

```python
usd = pd.read_csv('data/DX-Y.NYB.csv').dropna()
usd.head()
```

```python
usd = usd[['Date', 'Close']]
usd.columns = ['date_added', 'usd']
usd.head()
```

```python
# x = pd.concat([SP, VIX], axis=1)
ms1 = pd.merge(ms1, usd, how='left', on='date_added')
ms1.usd = ms1.usd.fillna(method='bfill')
ms1.usd = ms1.usd.fillna(method='ffill')
ms1
```
